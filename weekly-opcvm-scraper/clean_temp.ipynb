```python
import pandas as pd
```

```python
df = pd.read_csv("downloaded_files.csv")
df
```

```python
df["download_date"] = pd.to_datetime(df["download_date"])
df
```

```python
# Step 1: Extract the date part and convert to datetime
df['date'] = pd.to_datetime(df['filename'].str.replace('.pdf', '', regex=False), format='%d-%m-%Y')

# Step 2: Sort by the datetime column in descending order
df_sorted = df.sort_values(by='date', ascending=False).reset_index(drop=True)
df_sorted
```

```python
df_sorted.to_csv("downloaded_files.csv")
```
